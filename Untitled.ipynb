```python
# Initialize lists to store location values
locations_df1 = []
locations_df2 = []

# Process df1
for index, row in df1.iterrows():
    location = row["locations"]
    if location == "-":
        # Take value from the riskAllele column
        location = row["riskAllele"]
    
    # If location starts with "chr", remove it
    if location.startswith("chr"):
        location = location[3:]
        location = location.split('-')[0]
        
    df1.at[index, "locations"] = location
    
    # Store valid locations
    locations_df1.append(location)

# Process df2 (same as df1)
for index, row in df2.iterrows():
    location = row["locations"]
    if location == "-":
        # Take value from the riskAllele column
        location = row["riskAllele"]
    
    # If location starts with "chr", remove it
    if location.startswith("chr"):
        location = location[3:]
        location = location.split('-')[0]
        
    df2.at[index, "locations"] = location
    
    # Store valid locations
    locations_df2.append(location)

# Convert to sets to compare unique locations
set_df1_locations = set(locations_df1)
set_df2_locations = set(locations_df2)

# Compare the locations
common_locations = set_df1_locations.intersection(set_df2_locations)

sorted(common_locations)
```

```python
def extract_beta(beta_str):
    if isinstance(beta_str, str):  # Check if value is a string
        if beta_str == "-":
            return np.nan
        parts = beta_str.split(" ")  # Split the string
        value = float(parts[0])  # Extract numeric part
        if "decrease" in beta_str:
            return -abs(value)  # Make negative if "decrease"
        return abs(value)  # Keep positive if "increase"
    return np.nan  # Return NaN if not a string

# Convert beta column
df1_OR["beta_numeric"] = df1_OR["beta"].apply(extract_beta)

# Calculate OR values
df1_OR["calculated_or"] = np.exp(df1_OR["beta_numeric"])

# Replace '-' in orValue with calculated OR
df1_OR["orValue"] = df1_OR["orValue"].replace("-", np.nan)  # Convert "-" to NaN
df1_OR["orValue"] = df1_OR["orValue"].fillna(df1_OR["calculated_or"])  # Replace NaN with calculated OR

# Drop temporary columns
df1_OR = df1_OR.drop(columns=["beta_numeric", "calculated_or"])
```

```python
def extract_beta(beta_str):
    if isinstance(beta_str, str):  # Check if value is a string
        if beta_str == "-":
            return np.nan  # Return NaN for "-" values
        parts = beta_str.split(" ")  # Split the string
        value = float(parts[0])  # Extract numeric part
        if "decrease" in beta_str:
            return -abs(value)  # Make negative if "decrease"
        return abs(value)  # Keep positive if "increase"
    return np.nan  # Return NaN if not a string

# Convert beta column
df2_OR["beta_numeric"] = df2_OR["beta"].apply(extract_beta)

# Calculate OR values (Only for rows with valid beta values)
df2_OR["calculated_or"] = np.exp(df2_OR["beta_numeric"].fillna(0))  # Fill NaN with 0 to avoid issues in exp

# Replace '-' in orValue with calculated OR
df2_OR["orValue"] = df2_OR["orValue"].replace("-", np.nan)  # Convert "-" to NaN
df2_OR["orValue"] = df2_OR["orValue"].fillna(df2_OR["calculated_or"])  # Replace NaN with calculated OR

# Drop temporary columns
df2_OR = df2_OR.drop(columns=["beta_numeric", "calculated_or"])

```

```python
def extract_numeric(value):
    if isinstance(value, str):  # Check if the value is a string
        try:
            # Extract the numeric part from the string
            return float(value.split()[0])  # Split by space and take the first part
        except ValueError:
            return np.nan  # Return NaN if it cannot be converted to a number
    return value  # Return the value as is if it is already a numeric type

# Apply function to the orValue column
df2_OR["orValue"] = df2_OR["orValue"].apply(extract_numeric)

# Print updated DataFrame
df2_OR
```

```python


# List of columns to drop
columns_to_drop = ["traitName", "efoTraits", "accessionId", "pubmedId", "beta", "mappedGenes", "ci", "riskFrequency"]

# Drop unnecessary columns from both dataframes
df1_filtered = df1_OR.drop(columns=columns_to_drop)
df2_filtered = df2_OR.drop(columns=columns_to_drop)

# Merge on the 'locations' column, keeping separate columns for each disease


# Group df1 and df2 by riskAllele and locations
df1_grouped = df1_filtered.groupby(["locations", "riskAllele"]).agg({
    "pValue": "min",  # Keep the most significant (smallest) p-value
    
    "orValue": "mean"  # Average OR value
}).reset_index()

df2_grouped = df2_filtered.groupby(["locations", "riskAllele"]).agg({
    "pValue": "min",
    
    "orValue": "mean"
}).reset_index()

# Merge the grouped dataframes on both "locations" and "riskAllele"
merged_df = df1_grouped.merge(df2_grouped, on=["locations", "riskAllele"], suffixes=("_disease1", "_disease2"))

# Display merged dataframe
merged_df
```

```python
#volcano plot shows The strength and direction of genetic associations.

df1_OR["log_or"] = np.log(df1_OR["orValue"])  # Log(OR)
df1_OR["neg_log10_p"] = -np.log10(df1_OR["pValue"])  # -log10(p-value)

# Volcano plot using orValue (log(OR) vs -log10(p-value))
plt.figure(figsize=(8, 6))
plt.scatter(df1_OR["log_or"], df1_OR["neg_log10_p"], c="blue", alpha=0.5)

# Add significance threshold line (p-value < 5e-8)
plt.axhline(y=-np.log10(5e-8), color="red", linestyle="--", label="p-value < 5e-8")

# Add effect size threshold (log(OR) = 0)
plt.axvline(x=0, color="black", linestyle="--", label="Effect Size (log(OR)) = 0")

# Add labels and title
plt.xlabel("Effect Size (log(OR))")
plt.ylabel("-log10(p-value)")
plt.title("Volcano Plot (OR vs p-value)", fontsize=14)
plt.legend()

# Show the plot
plt.show()
```
